```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Send a GET request to the URL
url = "https://www.statista.com/statistics/1464487/india-most-polluted-states-by-pm10-level/"
response = requests.get(url)

# Step 2: Parse the HTML content of the page using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Step 3: Find the table in the webpage (assuming there is a table with relevant data)
# For Statista, the tables might be within divs or other elements, inspect the page to confirm
table = soup.find('table')

# Step 4: Extract the rows from the table
rows = table.find_all('tr')

# Step 5: Extract column headers (if available)
headers = [th.text.strip() for th in rows[0].find_all('th')]

# Step 6: Extract table data
data = []
for row in rows[1:]:
    columns = row.find_all('td')
    if columns:  # Only process rows with data
        data.append([col.text.strip() for col in columns])

# Step 7: Convert the data into a Pandas DataFrame
df = pd.DataFrame(data, columns=headers)

# Step 8: Display the DataFrame or save it to a file
print(df)

# Save to CSV if needed
df.to_csv('polluted_states_by_pm10.csv', index=False)
```
