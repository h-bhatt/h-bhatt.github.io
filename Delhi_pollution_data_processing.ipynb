```python
!pip install pandas

import pandas as pd

# Load pollution data
pollution_csv_path = "https://raw.githubusercontent.com/h-bhatt/h-bhatt.github.io/refs/heads/main/project_data/Delhi.csv"
pollution_data = pd.read_csv(pollution_csv_path)

# Convert date columns to datetime for proper handling
# Specify the correct format and dayfirst
pollution_data['Date'] = pd.to_datetime(pollution_data['Date'], format='%d/%m/%Y', dayfirst=True)

# Aggregate data for all stations
# Correct column name to "Date" as specified in the dataframe
aggregated_data = pollution_data.groupby(['Date']).mean().reset_index()
```
